```python
import gradio as gr
import re
from collections import Counter
import nltk
nltk.download('stopwords', quiet=True)
from nltk.corpus import stopwords

print("✅ AI Study Buddy LOADED SUCCESSFULLY!")
```

```python
stop_words = set(stopwords.words('english'))

def text_summarizer(text):
    """Smart extractive summarizer"""
    sentences = re.split(r'[.!?]+', text)
    sentences = [s.strip() for s in sentences if len(s.strip()) > 15]

    if len(sentences) < 2:
        return "Text too short for summary."

    # Score sentences by word importance
    words = re.findall(r'\b[a-zA-Z]{3,}\b', text.lower())
    word_freq = Counter(w for w in words if w not in stop_words)

    scores = []
    for sent in sentences:
        sent_words = re.findall(r'\b[a-zA-Z]{3,}\b', sent.lower())
        score = sum(word_freq.get(w, 0) for w in sent_words) / max(len(sent_words), 1)
        scores.append((sent, score))

    # Top 2-3 sentences
    top_sentences = sorted(scores, key=lambda x: x[1], reverse=True)[:3]
    return '. '.join(sent for sent, _ in top_sentences) + '.'

def concept_simplifier(topic):
    """Simple explanations"""
    templates = {
        'photosynthesis': "Plants use sunlight + water + CO2 to make food (glucose) and release oxygen.",
        'heart': "Heart is a muscle pump that sends blood to all body parts through arteries.",
        'gravity': "Gravity pulls objects toward Earth, keeping us grounded.",
        'electricity': "Electricity is energy flow through wires that powers devices.",
        'dna': "DNA is genetic code in cells that controls growth and traits.",
        'photosynthesis': "Plants make food from sunlight using green chlorophyll.",
        'mitochondria': "Powerhouse of cell that makes energy from food.",
        'chlorophyll': "Green chemical in plants that captures sunlight."
    }

    topic_key = topic.lower().strip()
    return templates.get(topic_key, f"'{topic}' is an important science concept.")

def extract_keywords(text, n=4):
    """Get important words"""
    words = re.findall(r'\b[a-zA-Z]{4,}\b', text.lower())
    filtered = [w for w in words if w not in stop_words]
    return [w.capitalize() for w, _ in Counter(filtered).most_common(n)]

def quiz_generator(text):
    """Create quiz questions"""
    keywords = extract_keywords(text)
    questions = []
    for i, keyword in enumerate(keywords):
        if i % 2 == 0:
            questions.append(f"Q{i+1}: Fill in blank: Plants need ______ for growth. ({keyword})")
        else:
            questions.append(f"Q{i+1}: Define '{keyword}' (1 sentence)")
    return "\n".join(questions)
```

```python
def process_input(action, text):
    if not text.strip():
        return "❌ Please enter text!"

    if action == "📝 Summarize":
        return text_summarizer(text)
    elif action == "🧠 Simplify":
        return concept_simplifier(text)
    elif action == "❓ Quiz":
        return quiz_generator(text)
    return "Select action above!"
```

```python
with gr.Blocks(title="AI Study Buddy") as demo:
    gr.Markdown("# 🤖 **AI Study Buddy**")
    gr.Markdown("*IBM SkillsBuild Internship Project*")
    gr.Markdown("*✅ ZERO ERRORS | ✅ INSTANT LOAD*")

    with gr.Row():
        with gr.Column(scale=2):
            input_text = gr.Textbox(
                label="📚 Study Material",
                placeholder="Paste notes or type topic...",
                lines=6
            )
            action = gr.Radio(
                choices=["📝 Summarize", "🧠 Simplify", "❓ Quiz"],
                value="📝 Summarize",
                label="Action"
            )
            btn = gr.Button("🚀 GENERATE", variant="primary")

        with gr.Column(scale=2):
            output = gr.Textbox(label="✨ Result", lines=10)

    # Examples
    gr.Examples(
        examples=[
            ["Photosynthesis is how plants make food using sunlight, water and carbon dioxide through chlorophyll."],
            ["photosynthesis"],
            ["Water cycle: evaporation, condensation, precipitation, collection."]
        ],
        inputs=[input_text]
    )

    # FIX: Event inside Blocks context
    btn.click(
        fn=process_input,
        inputs=[action, input_text],
        outputs=output
    )
```

```python
print("🎉 LAUNCHING...")
demo.launch(share=True, debug=True)
```
